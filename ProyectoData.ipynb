### Ahora, vamos a filtrar el artista de oro.

Vamos a intentar analizar más profundamente los datos. Quiero averiguar por qué estas canciones han llegado al top y quién es el artista afortunado de haber tenido más éxitos en el 2019. Para ello seguiremos los siguientes pasos

```python
#ahora que hemos visto los datos relevantes de la lista, vamos a filtrar datos. En este caso
# me interesa saber si se repite algún artista en la lista.
duplicate_values = lista50['Artist.Name'].duplicated(keep=False)
df_duplicated = lista50[duplicate_values] #preguntar pq no entendi
df_duplicated #df es data frame
```

```python
#Aqui se muestran una lista con los nombres en orden alfabeticoy la cantidad de veces que ha estado un artista en el top
df_duplicadosalfabeticamente = df_duplicated.sort_values(by="Artist.Name", ascending=True)
df_duplicadosalfabeticamente["Artist.Name"] 
```

```python
# este comando no lo necesito, pero realiza la media ordenados según el nombre de los artistas, de todos los 
# datos númericos de la tabla
lista50.groupby('Artist.Name').mean( )
```

A continuación, creamos una barplot para visualizar numericamente cuantas veces se repite el nombre de un artista en el data frame de nombres de artistas.

```python
# estos metodos son de la libreria seaborn. 
p = sb.histplot(data=df_duplicadosalfabeticamente, y='Artist.Name', hue = 'Artist.Name', shrink=0.9)

# el histplot es un hitograma y lo uso porque la cantidad de veces que se repite el nombre de
# un artista es una variable continua
sb.move_legend(p, "upper left", bbox_to_anchor=(1, 1))
```

Como se puede observar en la tabla, el artista Ed Sheeran fue el que mayor número de canciones tenía en el top 50, siendo el único que tuvo más de 2 éxitos, en su caso tuvo 4, como podemos observar en el eje x de la tabla
